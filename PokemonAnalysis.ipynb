### data visualisation 

```python
plt.figure(figsize=(5,3),dpi=150)
plt.hist(df['Speed'],color='green')
plt.title("speed distribution of pokemons",fontdict={'font':'consolas','fontsize':12})
plt.xlabel('speed (km/hr)')
plt.ylabel('No of pokemons')
plt.show()
```

```python
#find the count of generations in table 
generations=df.groupby(['Generation']).size()
generations

#create a pie chart for distribution

labels=[i for i in range(1,7)]
plt.pie(generations ,labels=labels,autopct='%.2f %%')
plt.title("generation distribution of pokemon",fontsize=12)
plt.show()
```

```python
generations=df.groupby(['Generation']).size()
plt.figure(figsize=(8,6),dpi=100)
plt.plot(generations,'r.--')
plt.ylabel("no of pokemons")
plt.yticks([i for i in range(10,200,20)])
plt.xlabel('Generation')
plt.show()
```

```python
#get the count of elements in a column
type1=df.groupby(df['Type 1']).size()

#set the plot style
plt.style.use('ggplot')

#get unique elements from a column
labels=df['Type 1'].unique()
explode=[0.2 for i in range(len(labels))]

#set the size of the figure 
plt.figure(dpi=200)

#plot using pie chart 
plt.pie(type1,labels=labels,explode=explode,textprops={'fontsize': 4},autopct='%.2f ')
plt.title("type 1 distribution")
plt.show()
```

```python
#create two df from dataset 
legendary=df.loc[df['Legendary']==True]['Speed']
NonLegend=df.loc[~df['Legendary']==True]['Speed']

#plt in box and whiskers chart 
plt.style.use('default')
labels=['legendary','Non legendary']
chart=plt.boxplot([legendary,NonLegend],labels=labels,patch_artist=True)
plt.ylabel("speed in km/hr")
plt.title("speed comparision of legendary vs non legendary pokemons")
colors=['red','blue']
for box,color in zip(chart['boxes'],colors):
    box.set(color=color,linewidth=3)

plt.show()
```
